```python
import seaborn as sns
import matplotlib.pyplot as plt
from adjustText import adjust_text
%matplotlib inline
```

```python
x = [0.144, 1.175, 0.156, 0.250, 0.200, 0.000, 0.250, 0.250]
y = [0, 0, 0, 0, 0, 0, 0, 0]
strings = ['O5-C1-C2-O2',
 'O5-C1-C2-O2',
 'O5-C1-C2-C3',
 'N1-C1-C2-C3',
 'N1-C1-C2-C3',
 'O5-C1-C2-H2',
 'O5-C1-C2-H2',
 'H1-C1-C2-O2']
```

```python
sns.set()
sns.set_context("talk")
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(6 * 1.2, 6))
plt.scatter(x, y)
texts = [plt.text(i, j, k, fontsize=14) for i, j, k in zip(x, y, strings)]

lims = (-0.05, 1.25)
plt.plot(lims, lims, "k-", lw=1, alpha=0.5, zorder=0)
plt.grid(zorder=-1)
plt.xlim(lims)
plt.ylim(lims)
adjust_text(texts, 
            arrowprops=dict(arrowstyle='-', color='gray', alpha=.5),
           precision=1e-6
            )
```
