```python
windows = [f"a{i:03d}" for i in range(14)] + ["p000"]
```

```python
path = "/home/davids4/gpfs/smirnoff-host-guest-simulations-data/systems/a-coc-p/smirnoff"
```

```python
distances = []
for index, window in enumerate(windows):
    this_window = os.path.join(path, window)
    trajectories = glob.glob(this_window + "/vac.*.nc")
    this_window_distances = []
    for trajectory in sorted(trajectories):
        print(trajectory)
        traj = pt.iterload(trajectory, top = this_window + "/vac.prmtop", stride=10)
        dist = pt.distance(traj, "@152 @127")
        this_window_distances.append(dist)
    distances.append(np.concatenate(this_window_distances).ravel())
```

```python
plt.figure()
for window in distances:
    plt.hist(window, bins=20, normed=True, alpha=0.2)
```
