```python
%load_ext autoreload
%autoreload 2

import os as os
import numpy as np
import glob
import shutil

from smirnovert.convert import convert
```

## For the confirmation, copy.

```python
reference_prmtop = 'systems/a-bam-p/original/*/original/full.hmr.topo'
reference_inpcrd = 'systems/a-bam-p/original/*/original/full.crds'

confirm_directory = 'systems/a-bam-p/confirm-original/'
smirnoff_directory = 'systems/a-bam-p/smirnoff/'
```

```python
windows = ['a000', 'a001', 'a002', 'a003', 'a004', 'a005', 'a006', 'a007', 'a008', 'a009', 'a010', 'a011', 'a012', 'a013', 'p000', 'p001', 'p002', 'p003', 'p004', 'p005', 'p006', 'p007', 'p008', 'p009', 'p010', 'p011', 'p012', 'p013', 'p014', 'p015', 'p016', 'p017', 'p018', 'p019', 'p020', 'p021', 'p022', 'p023', 'p024', 'p025', 'p026', 'p027', 'p028', 'p029', 'p030', 'p031', 'p032', 'p033', 'p034', 'p035', 'p036', 'p037', 'p038', 'p039', 'p040', 'p041', 'p042', 'p043', 'p044', 'p045']

for window in windows:
    try:
        os.mkdir(confirm_directory + window)
        os.mkdir(smirnoff_directory + window)
    except:
        pass
```

```python
for inpcrd, prmtop in zip(sorted(glob.glob(reference_inpcrd)), sorted(glob.glob(reference_prmtop))):

    window = prmtop.split('/')[3]
    phase = window[0]
    window_padded = '{:03d}/'.format(int(window[1:]))
    
    if phase == 'u':
        phase = 'p'
    
    shutil.copy(inpcrd, confirm_directory + phase + window_padded)
    shutil.copy(prmtop, confirm_directory + phase + window_padded)
```

## For SMIRNOFF99Frosst, convert.

To make life easier, and allow us to use the `convert` function, let's copy the original files to the destination directory.

```python
for inpcrd, prmtop in zip(sorted(glob.glob(reference_inpcrd)), sorted(glob.glob(reference_prmtop))):

    window = prmtop.split('/')[3]
    phase = window[0]
    window_padded = '{:03d}'.format(int(window[1:]))
    
    if phase == 'u':
        phase = 'p'
    
    shutil.copy(inpcrd, smirnoff_directory + phase + window_padded)
    shutil.copy(prmtop, smirnoff_directory + phase + window_padded)
    
    destination_path = smirnoff_directory + phase + window_padded + '/'
    prmtop = os.path.basename(prmtop)
    inpcrd = os.path.basename(inpcrd)
    
    print(destination_path)
    
    convert(destination=destination_path,
           prefix='smirnoff',
           reference_prmtop=prmtop,
           reference_inpcrd=inpcrd,
           host_resname='MGO',
           guest_resname='BAM',
           dummy=True,
           debug=False)
```
