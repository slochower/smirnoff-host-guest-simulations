````python
%load_ext autoreload
%autoreload 2

import os as os
import numpy as np

import parmed as pmd
import pytraj as pt

from smirnovert import utils
````

````python
reference = pmd.load_file('systems/a-bam-p/original/a00/original/full.hmr.topo',
                         structure=True)
````

````python
smirnoff =  pmd.load_file('systems/a-bam-p/smirnoff/a000/smirnoff.prmtop',
                         structure=True)
````

````python
reference_types = []
for atom in reference.atoms:
    if atom.type not in reference_types:
        reference_types.append(atom.type)
    else:
        pass
print(reference_types)
````

````python
reference_lj = {}

for atom in reference.atoms:
    if atom.type not in reference_lj:
        reference_lj[atom.type] = {
            'sigma' : atom.sigma,
            'epsilon' : atom.epsilon
        }
    else:
        pass
````

````python
reference_lj
````

````python
smirnoff_types = []
for atom in smirnoff.atoms:
    if atom.type not in smirnoff_types:
        smirnoff_types.append(atom.type)
    else:
        pass
print(smirnoff_types)

smirnoff_lj = {}

for atom in smirnoff.atoms:
    if atom.type not in smirnoff_lj:
        smirnoff_lj[atom.type] = {
            'sigma' : atom.sigma,
            'epsilon' : atom.epsilon
        }
    else:
        pass
````

Use mapping to show association and then group together.

````python
reference.save('reference.mol2', overwrite=True)
smirnoff.save('smirnoff.mol2', overwrite=True)
````

````python
reference_oemol = utils.load_mol2('reference.mol2')
smirnoff_oemol = utils.load_mol2('smirnoff.mol2')
````

````python
mapping = utils.map_atoms(reference_oemol, smirnoff_oemol)
````

````python
for key, value in mapping.items():
    print(f'{reference.atoms[key].atom_type} → {smirnoff.atoms[value].atom_type}')
````

````python
reduced_mapping = {}
excluded_atoms = ['HW', 'OW', 'Cl-', 'Na+', 'Pb']

for key, value in mapping.items():
    if str(reference.atoms[key].atom_type) not in reduced_mapping and \
    str(reference.atoms[key].atom_type) not in excluded_atoms:
        reduced_mapping[str(reference.atoms[key].atom_type)] = []
    if str(reference.atoms[key].atom_type) not in excluded_atoms:
        reduced_mapping[str(reference.atoms[key].atom_type)].append(str(smirnoff.atoms[value].atom_type))
````

````python
reduced_mapping
````

````python
for key, values in reduced_mapping.items():
    smirnoff_sigmas = set([smirnoff_lj[value]['sigma'] for value in values])
    print(f"{key} → {values}")
    print(f"{reference_lj[key]['sigma']} → {smirnoff_sigmas}")
````

````python
print('Rmin/2')
for key, values in reduced_mapping.items():
    smirnoff_sigmas = set([smirnoff_lj[value]['sigma']*2**(1./6.)/2.0  for value in values])
    print(f"{key} → {values}")
    print(f"{reference_lj[key]['sigma']*2**(1./6.)/2.0} → {smirnoff_sigmas}")
````
